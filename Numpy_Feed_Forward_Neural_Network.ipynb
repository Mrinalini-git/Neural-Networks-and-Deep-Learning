```python
import numpy as np

# New input values
x = np.array([1.0, 0.5, 0.3])

# New weight matrix
W = np.array([
    [0.2, 0.4],
    [0.7, 0.9],
    [0.5, 0.2]
])

# New bias for hidden layer
b1 = np.array([0.05, 0.03])

# Calculate input to hidden layer
z1 = np.dot(x, W) + b1

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Apply sigmoid activation to hidden layer
h = sigmoid(z1)
print("Hidden layer output (h):", h)

# New bias for output layer
b2 = 0.02

# Calculate input to output layer (sum of hidden activations + bias)
z2 = np.sum(h) + b2

# Apply sigmoid activation to output layer
y_pred = sigmoid(z2)
print("Predicted output (y_pred):", y_pred)

# Target value
y = 0.8

# Calculate Mean Squared Error (MSE) loss
loss = np.mean((y_pred - y)**2)
print("Loss (MSE):", loss)
```
