```python
import numpy as np

x = np.array([0.5, 0.2, 0.3])
w = np.array([[0.2, 0.1], [0.3, 0.5] , [0.6, 0.4]])
b = np.array([0.1, 0.2])

Z = np.dot(x, w) + b

sigmoid = 1 / (1 + np.exp(-Z))

print("Activated Output")
bias1 = sigmoid[0] * 0.2
bias2 = sigmoid[1] * 0.3
print(bias1)
print(bias2)
```

```python
x = np.array([0.5, 0.2, 0.3])
w = np.array([[0.2, 0.1], [0.3, 0.5] , [0.6, 0.4]])
b = np.array([0.1, 0.2])

Z = np.dot(x, w) + b

relu = np.maximum(0, Z)

print("ReLU Output:")
print(relu)

print("Activated Output")
weight1 = relu[0] * 0.2
weight2 = relu[1] * 0.3
print(weight1)
print(weight2)
```

```python
y_target = np.array([0.5, 0.6])
mse_loss = np.mean((Z - y_target)**2)

print("Target Output (y_target):")
print(y_target)
print("Mean Squared Error (MSE):")
print(mse_loss)
```

```python
dZ_loss = Z - y_target
dw = np.outer(x, dZ_loss)

print("Derivative of loss with respect to Z (dZ_loss):")
print(dZ_loss)
print("Gradients for weights (dw):")
print(dw)
```

```python
db = dZ_loss

print("Gradients for biases (db):")
print(db)
```
